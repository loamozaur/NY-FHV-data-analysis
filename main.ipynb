```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import math
```

# Preprocessing

```python
cols = ['pickup_datetime', 'dropoff_datetime', 'PULocationID', 'DOLocationID']
data = pd.read_csv('fhvhv.csv', usecols = cols, parse_dates = [0, 1])
data.tail(15)
```

```python
data = data.rename(columns={'pickup_datetime': 'PUdt', 'dropoff_datetime': 'DOdt', 'PULocationID': 'PUloc', 'DOLocationID': 'DOloc'})
data
```

# Pickup and Dropoff Pattern

Creating pickup and dropoff hours and weekdays columns

```python
data[['PUhour', 'DOhour']] = data[['PUdt', 'DOdt']].apply(lambda x: x.dt.hour)
data[['PUday', 'DOday']] = data[['PUdt', 'DOdt']].apply(lambda x: x.dt.weekday)
data.tail()
```

```python
PUcols = ['PUloc', 'PUhour', 'PUday']
PUdata_by_hour = data[PUcols].groupby(['PUloc', 'PUhour']).size().reset_index(name='count')
PUdata_by_hour
```

```python
PUdata_by_hour.nlargest(10, 'count')
```

```python
DOcols = ['DOloc', 'DOhour', 'DOday']
DOdata_by_hour = data[DOcols].groupby(['DOloc', 'DOhour']).size().reset_index(name='count')
DOdata_by_hour
```

```python
DOdata_by_hour.nlargest(10, 'count')
```

```python
PUdata_by_day = data[PUcols].groupby(['PUloc', 'PUday']).size().reset_index(name='count')
PUdata_by_day
```

```python
PUdata = data[PUcols].groupby(['PUday']).size().reset_index(name='count')
PUdata = PUdata.set_index('PUday').apply(lambda x: (x/2))
PUdata.nlargest(10, 'count')
```

```python
DOdata_by_day = data[DOcols].groupby(['DOloc', 'DOday']).size().reset_index(name='count')
DOdata_by_day
```

```python
DOdata = data[DOcols].groupby(['DOday']).size().reset_index(name='count')
DOdata = DOdata.set_index('DOday').apply(lambda x: (x/2)) #Because the data is for 2 weeks
DOdata.nlargest(7, 'count')
```

```python
weekdays = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

plt.style.use('_classic_test')
plt.bar(DOdata.index - 0.2, DOdata['count']/1000, color='green', width=0.4, label='Dropoff')
plt.bar(PUdata.index + 0.2, PUdata['count']/1000, color='yellow', width=0.4, label='Pick-up')
plt.legend(loc='upper left')
plt.title('Counts by weekday')
plt.xlabel('Day of Week')
plt.ylabel('1,000 trips per unit')
plt.xticks(ticks=np.arange(7), labels=weekdays)
plt.show()
```

```python
PUcols1 = ['PUloc', 'PUhour']
PUdata1 = data[PUcols1].groupby(['PUhour']).size().reset_index(name='count')
PUdata1 = PUdata1['count']/14
PUdata1
```

```python
DOcols1 = ['DOloc', 'DOhour']
DOdata1 = data[DOcols1].groupby(['DOhour']).size().reset_index(name='count')
DOdata1 = DOdata1['count']/14
DOdata1
```

```python
plt.style.use('_classic_test')
plt.plot(DOdata1.index + 0.5, DOdata1/1000, color='green', linewidth=3, label='Dropoff')
plt.plot(PUdata1.index + 0.5, PUdata1/1000, color='pink', linewidth=3, label='Pick-up')
plt.title('Trips by hour of day')
plt.xlabel('hour')
plt.ylabel('1,000 trips per unit')
plt.legend(loc='upper left')
plt.xlim([0, 24])
plt.grid(which='both', alpha=1)
plt.show()
```

Top areas for pick-up and dropoff location ID.
Group of location IDs is called an area. 
Location IDs have been assigned manually to areas.

```python
cols = ['LocationID', 'areaID']
zones = pd.read_csv('zones.csv', usecols =cols)
zones
```

```python
PUtop = data.groupby('PUloc').size().reset_index(name='count')
PUtop.nlargest(10,'count')
```

```python
DOtop = data.groupby('DOloc').size().reset_index(name='count')
DOtop.nlargest(10, 'count')
```

```python
cols = ['PUloc', 'DOloc', 'PUhour', 'PUday']
OD = pd.merge(data[cols], zones, how='left', left_on=['PUloc'], right_on=['LocationID']).drop('LocationID', axis=1)
OD
```

```python
# for pickup
ODcount = OD['areaID'].value_counts()/14
ODindex = ODcount.head(10).index
ODvalues = ODcount.head(10).values
ODindex
```

```python
plt.bar(ODindex, ODvalues/1000, color='green')
plt.title('Number of pick-ups by areas')
plt.xlabel('Area ID')
plt.ylabel('in 1,000 trips per hour')
plt.show()
```
